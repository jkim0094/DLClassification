<span style="color:red">**(a)**</span> Exponential linear unit (ELU): $\text{ELU}(x)=\begin{cases}
0.1\left(\exp(x)-1\right) & \text{if}\,x\leq0\\
x & \text{if}\,x>0
\end{cases}$
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div>

<span style="color:red">**(b)**</span> Gaussian Error Linear Unit (GELU): $\text{GELU}(x)=x\Phi(x)$ where $\Phi(x)$ is the `probability cummulative function` of the standard Gaussian distribution or $\Phi(x) = \mathbb{P}\left(X\leq x\right)$ where $X \sim N\left(0,1\right)$. In addition, the GELU activation fuction (the link for the [main paper](https://arxiv.org/pdf/1606.08415v5.pdf)) has been widely used in the state-of-the-art Vision for Transformers (e.g., here is the link for [the main ViT paper](https://arxiv.org/pdf/2010.11929v2.pdf)).
<div style="text-align: right"><span style="color:red">[1.5 points]</span></div>

### (a)
1. Output range: $(-0.1, \infty)$

2. Derivative:

- For $x > 0$:  
  $$
  \frac{d}{dx}[ELU(x)] = \frac{d}{dx}[x] = 1
  $$

- For $x \leq 0$:  
  $$
  \frac{d}{dx}[ELU(x)] = \frac{d}{dx}[0.1(\exp(x) - 1)] = 0.1 \cdot \exp(x)
  $$

Therefore, the derivative of ELU is:

$$
ELU'(x) =
\begin{cases}
0.1 \cdot \exp(x) & \text{if } x \leq 0 \\
1 & \text{if } x > 0
\end{cases}
$$


### (b)
1. Output range: $(-0.17, \infty)$

2. Derivative:

GELU :
$$
\text{GELU}(x) = x \cdot \Phi(x)
$$
where $\Phi(x)$ is the cumulative distribution function (CDF):
$$
\Phi(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}} e^{-t^2 / 2} \, dt
$$
To make computations more efficient, GELU can be approximated as follows:
$$
\text{GELU}(x) \approx 0.5x \left( 1 + \tanh \left( \sqrt{\frac{2}{\pi}} \left( x + 0.044715x^3 \right) \right) \right)
$$

The derivative of the approximated GELU function is given by:
$$
\text{GELU}'(x) = 0.5 \left( 1 + \tanh \left( \sqrt{\frac{2}{\pi}} \left( x + 0.044715x^3 \right) \right) \right)
+ 0.5x \left( 1 - \tanh^2 \left( \sqrt{\frac{2}{\pi}} \left( x + 0.044715x^3 \right) \right) \right)
\cdot \sqrt{\frac{2}{\pi}} \left( 1 + 3 \cdot 0.044715x^2 \right)
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define ELU function
def elu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Define ELU derivative function
def elu_derivative(x, alpha=0.1):
    return np.where(x > 0, 1, alpha * np.exp(x))

# Set range for x values
x = np.linspace(-3, 3, 400)

# Remove x = 0 values
x_no_zero = np.linspace(-3, 3, 400)
x_no_zero = x_no_zero[x_no_zero != 0]  # Remove x = 0

# Calculate ELU and ELU derivative
y_elu = elu(x)
y_elu_derivative = elu_derivative(x_no_zero)

# Plot the graphs
plt.figure(figsize=(12, 6))

# Plot ELU function
plt.subplot(1, 2, 1)
plt.plot(x, y_elu, label='ELU(x)')
plt.title('ELU Activation Function')
plt.xlabel('x')
plt.ylabel('ELU(x)')
plt.grid(True)
plt.legend()

# Plot ELU derivative
plt.subplot(1, 2, 2)
plt.scatter(x_no_zero, y_elu_derivative, label="ELU'(x)", color='red', s=10)  # Exclude x = 0 and plot as dots
plt.title('Derivative of ELU Activation Function')
plt.xlabel('x')
plt.ylabel("ELU'(x)")
plt.grid(True)
plt.legend()

# Add special points at x = 0
plt.scatter(0, 0.1, color='red', edgecolor='black', zorder=5)  # Point from left side
plt.scatter(0, 1, color='white', edgecolor='black', zorder=5)  # Point from right side

plt.tight_layout()
plt.show()

# Define GELU function
def gelu(x):
    return x * norm.cdf(x)

# Define GELU derivative function
def gelu_derivative(x):
    pdf = norm.pdf(x)  # PDF of standard normal distribution
    return norm.cdf(x) + x * pdf

# Set range for x values
x = np.linspace(-5, 5, 400)

# Calculate GELU and GELU derivative
y_gelu = gelu(x)
y_gelu_derivative = gelu_derivative(x)

# Plot the graphs
plt.figure(figsize=(12, 6))

# Plot GELU function
plt.subplot(1, 2, 1)
plt.plot(x, y_gelu, label='GELU(x)')
plt.title('GELU Activation Function')
plt.xlabel('x')
plt.ylabel('GELU(x)')
plt.grid(True)
plt.legend()

# Plot GELU derivative
plt.subplot(1, 2, 2)
plt.plot(x, y_gelu_derivative, label="GELU'(x)", color='red')
plt.title('Derivative of GELU Activation Function')
plt.xlabel('x')
plt.ylabel("GELU'(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### (f)
The cross-entropy loss :
$$
L = -\sum_{i} y_i \log(p_i)
$$

Here, the correct class is 1 \(y = 2\), and all other classes are 0 (y = 1\).

Given the probability distribution:
$$
p(x) = \begin{bmatrix} 1.41 \times 10^{-16} & 1 \end{bmatrix}
$$

The calculation of the loss is as follows:

$$
\begin{align*}
L &= -(0 \cdot \log(1.41 \times 10^{-16}) + 1 \cdot \log(1)) \\
L &= -(\log(1)) \\
L &= 0
\end{align*}
$$

The cross-entropy loss \(L = 0\), which indicates that the model's prediction perfectly matches the actual label.

### (g)
**Why the Cross-Entropy Loss is Never Negative:**

The cross-entropy loss is never negative because the probability \(p\) is always between 0 and 1. When we take the logarithm of a number in this range, the result is always a negative value. However, since the cross-entropy formula includes a negative sign in front, the overall value becomes positive or zero.

**When Cross-Entropy Loss is Zero:**

The cross-entropy loss becomes zero when the predicted probability for the correct class is 1. This happens when the model's prediction perfectly matches the actual label.

```python
class MyLinear(torch.nn.Module):
  def __init__(self, input_size, output_size):
    """
    input_size: the size of the input
    output_size: the size of the output
    """
    super().__init__()
    #Your code here
    self.input_size = input_size
    self.output_size = output_size
    self.W = torch.nn.Parameter(torch.nn.init.xavier_uniform_(torch.empty(self.output_size, self.input_size)))
    self.b = torch.nn.Parameter(torch.zeros(self.output_size))
  #forward propagation
  def forward(self, x): #x is a mini-batch
    #Your code here
    return x@self.W.T + self.b


```

```python
class MyFFN(torch.nn.Module):
  def __init__(self, input_size, num_classes, hidden_sizes, act = torch.nn.ReLU()):
    """
    input_size: the size of the input
    num_classes: the number of classes
    act is the activation function
    hidden_sizes is the list of hidden sizes
    for example input_size = 3, hidden_sizes = [5, 7], num_classes = 4, and act = torch.nn.ReLU()
    means that we are building up a FFN with the confirguration
    (3 (Input) -> 5 (ReLU) -> 7 (ReLU) -> 4 (Output))
    """
    super(MyFFN, self).__init__()
    self.input_size = input_size
    self.num_classes = num_classes
    self.act = act
    self.hidden_sizes = hidden_sizes
    self.num_layers = len(hidden_sizes) + 1

  def create_FFN(self):
    """
    This function creates the feed-forward neural network
    We stack many MyLinear layers
    """
    hidden_sizes = [self.input_size] + self.hidden_sizes + [self.num_classes]
    self.layers = []
    #Your code here
    # Loop through hidden sizes to create layers
    for i in range(len(hidden_sizes) - 1):
        layer = MyLinear(hidden_sizes[i], hidden_sizes[i+1])
        self.layers.append(layer)

    # Convert layers to torch.nn.ModuleList
    self.layers = torch.nn.ModuleList(self.layers)

  def forward(self,x):
    """
    This implements the forward propagation of the batch x
    This needs to return the prediction probabilities of x
    """
    #Your code here
    # Pass input x through layer and activation except for last
    for i in range(len(self.layers) - 1):
        x = self.act(self.layers[i](x))

    # Last layer without activation
    x = self.layers[-1](x)

    # Return softmax probabilities
    return torch.nn.functional.softmax(x, dim=1)

  def compute_loss(self, x, y):
    """
    This function computes the cross-entropy loss
    You can use the built-in CE loss of PyTorch
    """
    #Your code here
    # prediction
    predictions = self.forward(x)

    # CE
    loss_fn = torch.nn.CrossEntropyLoss()

    # Loss
    loss = loss_fn(predictions, y)

    return loss
  def update_SGD(self, x, y, learning_rate = 0.01):
    """
    This function updates the model parameters using SGD using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    # Loss
    loss = self.compute_loss(x, y)

    # backpropagation
    loss.backward()

    # SGD
    with torch.no_grad():   # Disable gradient tracking
        for layer in self.layers:
          # SGD formular
          layer.W -= learning_rate * layer.W.grad
          layer.b -= learning_rate * layer.b.grad

          # Reset gradients to zero for next batch
          layer.W.grad.zero_()
          layer.b.grad.zero_()

  def update_SGDwithMomentum(self, x, y, learning_rate = 0.01, momentum = 0.9):
    """
    This function updates the model parameters using SGD with momentum using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    # Loss
    loss = self.compute_loss(x, y)

    # backpropagation
    loss.backward()

    # SGD w Momentum
    with torch.no_grad():
        # Ditionary for storing momentum
        if not hasattr(self, 'velocities'):
            self.velocities = {}

        for i, layer in enumerate(self.layers):
            if isinstance(layer, MyLinear):
                # Initialize momentum velocities if not already done
                if i not in self.velocities:
                    self.velocities[i] = {
                        'W': torch.zeros_like(layer.W),
                        'b': torch.zeros_like(layer.b)
                    }

                # SGD w Momentum formualr
                self.velocities[i]['W'] = momentum * self.velocities[i]['W'] + learning_rate * layer.W.grad
                self.velocities[i]['b'] = momentum * self.velocities[i]['b'] + learning_rate * layer.b.grad


                layer.W -= self.velocities[i]['W']
                layer.b -= self.velocities[i]['b']


                layer.W.grad.zero_()
                layer.b.grad.zero_()

  def update_AdaGrad(self, x, y, learning_rate = 0.01):
    """
    This function updates the model parameters using AdaGrad using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    # Loss
    loss = self.compute_loss(x, y)

    # Backpropagaiton
    loss.backward()

    # AdaGrad
    with torch.no_grad():
        # Ditionary for storing momentum
        if not hasattr(self, 'squared_gradients'):
            self.squared_gradients = {}

        for i, layer in enumerate(self.layers):

            # Initialize squared gradients if not already done
            if i not in self.squared_gradients:
                self.squared_gradients[i] = {
                    'W': torch.zeros_like(layer.W),
                    'b': torch.zeros_like(layer.b)
                }

            # Accumulate squared gradients
            self.squared_gradients[i]['W'] += layer.W.grad ** 2
            self.squared_gradients[i]['b'] += layer.b.grad ** 2

            # AdaGrad formualr
            layer.W -= learning_rate * layer.W.grad / (torch.sqrt(self.squared_gradients[i]['W']) + 1e-7)
            layer.b -= learning_rate * layer.b.grad / (torch.sqrt(self.squared_gradients[i]['b']) + 1e-7)

            # Reset gradients to zero
            layer.W.grad.zero_()
            layer.b.grad.zero_()
```

We now develop a feed-forward neural network with the architecture $784 \rightarrow 40(ReLU) \rightarrow 30(ReLU) \rightarrow 10(softmax)$.
```python
import torch
import torch.nn as nn
import torch.optim as optim

# Define neural network class (using Sequential)
class DNN(nn.Module):
    def __init__(self, n1=40, n2=30, act_fn=nn.ReLU()):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(784, n1),  # Input: 784, Output: n1
            act_fn,              # Activation function for the first hidden layer
            nn.Linear(n1, n2),   # Input: n1, Output: n2
            act_fn,              # Activation function for the second hidden layer
            nn.Linear(n2, 10),   # Output: 10 classes
            nn.Softmax(dim=1)    # Apply softmax to output layer
        )

    def forward(self, x):
        return self.network(x)  # Perform forward pass through the Sequential layers

# Define training loop function
def train_model(model, optimizer, criterion, train_loader, device):
    model.train()  # Set model to training mode
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)  # Move data to GPU
        images = images.view(-1, 28*28)  # Reshape images

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimizer update
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calculate loss and accuracy
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # Return average loss and accuracy per epoch
    train_loss = running_loss / len(train_loader)
    train_acc = 100 * correct / total
    return train_loss, train_acc

# Define evaluation function
def evaluate_model(model, criterion, test_loader, device):
    model.eval()  # Set model to evaluation mode
    test_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            images = images.view(-1, 28*28)

            # Forward pass and loss calculation
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()

            # Calculate accuracy
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    test_loss /= len(test_loader)
    test_acc = 100 * correct / total
    return test_loss, test_acc

# Initialize model and set training/evaluation settings
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DNN().to(device)  # Move model to GPU
criterion = nn.CrossEntropyLoss()  # Loss function
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam optimizer
num_epochs = 50

# Run training and evaluation
for epoch in range(num_epochs):
    train_loss, train_acc = train_model(model, optimizer, criterion, train_loader, device)
    test_loss, test_acc = evaluate_model(model, criterion, test_loader, device)  # Added device

    print(f"Epoch [{epoch+1}/{num_epochs}], "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, "
          f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%")
```

####  <span style="color:red">**Question 2.5:**</span> Experimenting with **sharpness-aware minimization** technique
<div style="text-align: right"> <span style="color:red">[5 points]</span> </div>

Sharpness-aware minimization (SAM) (i.e., [link for main paper](https://openreview.net/pdf?id=6Tm1mposlrM) from Google Deepmind) is a simple yet but efficient technique to improve the generalization ability of deep learning models on unseen data examples. In your research or your work, you might potentially use this idea. Your task is to read the paper and implement *Sharpness-aware minimization (SAM)*. Finally, you need to apply SAM to the best architecture found in **Question 2.3**.   

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import Optimizer

# SAM optimizer implementation
class SAM(Optimizer):
    def __init__(self, params, base_optimizer, rho=0.05, **kwargs):
        assert rho >= 0.0, f"Invalid rho, should be non-negative: {rho}"
        defaults = dict(rho=rho, **kwargs)
        super(SAM, self).__init__(params, defaults)
        self.base_optimizer = base_optimizer(self.param_groups, **kwargs)
        self.param_groups = self.base_optimizer.param_groups

    @torch.no_grad()
    def first_step(self, zero_grad=False):
        grad_norm = self._grad_norm()  # Calculate gradient norm
        for group in self.param_groups:
            scale = group["rho"] / (grad_norm + 1e-12)

            for p in group["params"]:
                if p.grad is None:
                    continue
                if p not in self.state:
                    self.state[p] = {}
                self.state[p]["old_p"] = p.data.clone()  # Save original parameter
                p.add_(p.grad, alpha=scale)  # Update parameter

        if zero_grad:
            self.zero_grad()  # Clear gradients

    @torch.no_grad()
    def second_step(self):
        for group in self.param_groups:
            for p in group["params"]:
                if p.grad is None:
                    continue
                p.data = self.state[p]["old_p"]  # Restore original parameter
        self.base_optimizer.step()  # Perform optimization step

    def _grad_norm(self):
        # Calculate L2 norm of the gradients
        norm = torch.norm(torch.stack([p.grad.norm(p=2) for group in self.param_groups for p in group["params"] if p.grad is not None]))
        return norm

    def step(self, closure=None):
        assert closure is not None, "SAM requires closure"
        closure = torch.enable_grad()(closure)

        # First optimization step
        self.first_step(zero_grad=True)
        closure()  # Recalculate gradients
        self.second_step()  # Second optimization step


# Define model and move to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DNN(n1=40, n2=40, act_fn=nn.Tanh()).to(device)

# Define SAM optimizer
base_optimizer = torch.optim.Adam
optimizer = SAM(model.parameters(), base_optimizer, rho=0.05, lr=0.0001)

# Loss function
criterion = nn.CrossEntropyLoss()

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    # Training phase
    for images, labels in train_loader:
        images = images.view(images.size(0), -1).to(device)  # Flatten images
        labels = labels.to(device)

        # Closure for SAM optimizer
        def closure():
            optimizer.zero_grad()  # Clear gradients
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()  # Backpropagation
            return loss

        loss = closure()  # Compute loss and backpropagation
        optimizer.first_step(zero_grad=True)  # First SAM step

        # Recalculate gradients for second SAM step
        closure()
        optimizer.second_step()  # Second SAM step

        # Calculate training loss and accuracy
        outputs = model(images)
        loss = criterion(outputs, labels)
        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    train_loss = running_loss / len(train_loader)
    train_acc = 100 * correct_train / total_train

    # Validation phase
    model.eval()
    valid_loss = 0.0
    correct_valid = 0
    total_valid = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images = images.view(images.size(0), -1).to(device)  # Flatten images
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            valid_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_valid += labels.size(0)
            correct_valid += (predicted == labels).sum().item()

    valid_loss /= len(test_loader)
    valid_acc = 100 * correct_valid / total_valid


    print(f"Epoch [{epoch+1}/{num_epochs}]")
    print(f"train loss= {train_loss:.4f} – train acc= {train_acc:.2f}% – valid loss= {valid_loss:.4f} – valid acc= {valid_acc:.2f}%")
```

**First, you need to implement the block of our CNN in the class `YourBlock`. You can ignore `use_skip` and `skip connection` for simplicity. However, you cannot earn full marks for this question.**

```python
class YourBlock(nn.Module):
    def __init__(self, in_feature_maps, out_feature_maps, drop_rate=0.2, batch_norm=True, use_skip=True):
        super(YourBlock, self).__init__()
        self.use_skip = use_skip

        # Define layers using ModuleList
        self.block = nn.ModuleList()

        # First Conv + ReLU
        self.block.append(nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=3, stride=1, padding=1))
        self.block.append(nn.ReLU(inplace=True))

        # Add BatchNorm layer if batch_norm is True
        if batch_norm:
            self.block.append(nn.BatchNorm2d(out_feature_maps))

        # Second Conv + ReLU
        self.block.append(nn.Conv2d(out_feature_maps, out_feature_maps, kernel_size=3, stride=1, padding=1))
        self.block.append(nn.ReLU(inplace=True))

        # Add BatchNorm again if batch_norm is True
        if batch_norm:
            self.block.append(nn.BatchNorm2d(out_feature_maps))

        # Add MaxPool + Dropout layers
        self.block.append(nn.MaxPool2d(kernel_size=2, stride=2, padding=0))
        self.block.append(nn.Dropout(p=drop_rate))

        # Define skip connection if use_skip is True
        if self.use_skip:
            self.skip_conv = nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=1, stride=2)

    def forward(self, x):
        out = x
        # Apply layers from ModuleList one by one
        for layer in self.block:
            out = layer(out)

        # Apply skip connection if use_skip is True
        if self.use_skip:
            skip_out = self.skip_conv(x)
            out += skip_out

        return out
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def compute_loss(model, loss_fn, loader):
  loss = 0
  # Set model to eval mode for inference
  model.eval()
  with torch.no_grad():  # No need to track gradients for validation
    for (batchX, batchY) in loader:
      # Move data to the same device as the model
      batchX, batchY = batchX.to(device).type(torch.float32), batchY.to(device).type(torch.long)
      loss += loss_fn(model(batchX), batchY)
  # Set model back to train mode
  model.train()
  return float(loss)/len(loader)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def compute_acc(model, loader):
    correct = 0
    totals = 0
    # Set model to eval mode for inference
    model.eval()
    for (batchX, batchY) in loader:
        # Move batchX and batchY to the same device as the model
        batchX, batchY = batchX.to(device).type(torch.float32), batchY.to(device)
        outputs = model(batchX)  # feed batch to the model
        totals += batchY.size(0)  # accumulate totals with the current batch size
        predicted = torch.argmax(outputs.data, 1)  # get the predicted class
        # Move batchY to the same device as predicted for comparison
        correct += (predicted == batchY).sum().item()
    return correct / totals
```

**Here is the code to train our model.**

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def fit(model= None, train_loader = None, valid_loader= None, optimizer = None,
        num_epochs = 50, verbose = True, seed= 1234):
  torch.manual_seed(seed)
  # Move the model to the device before initializing the optimizer
  model.to(device) # Move the model to the GPU

  if optimizer == None:
    optim = torch.optim.Adam(model.parameters(), lr = 0.001) # Now initialize optimizer with model on GPU
  else:
    optim = optimizer
  history = dict()
  history['val_loss'] = list()
  history['val_acc'] = list()
  history['train_loss'] = list()
  history['train_acc'] = list()

  for epoch in range(num_epochs):
    model.train()
    for (X, y) in train_loader:
      # Move input data to the same device as the model
      X,y = X.to(device), y.to(device)
      # Forward pass
      outputs = model(X.type(torch.float32)) # X is already on the correct device
      loss = loss_fn(outputs, y.type(torch.long))
      # Backward and optimize
      optim.zero_grad()
      loss.backward()
      optim.step()
      #losses and accuracies for epoch
    val_loss = compute_loss(model, loss_fn, valid_loader)
    val_acc = compute_acc(model, valid_loader)
    train_loss = compute_loss(model, loss_fn, train_loader)
    train_acc = compute_acc(model, train_loader)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    if not verbose: #verbose = True means we do show the training information during training
      print(f"Epoch {epoch+1}/{num_epochs}")
      print(f"train loss= {train_loss:.4f} - train acc= {train_acc*100:.2f}% - valid loss= {val_loss:.4f} - valid acc= {val_acc*100:.2f}%")
  return history
```

```python
import torch.optim as optim
import numpy as np

# mixup data
def mixup_data(x, y, alpha=1.0):
    # If alpha is greater than 0, sample lambda from a Beta distribution
    # If alpha is 0 or less, set lambda to 1
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1.0

    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device) # Create a random index to shuffle the batch
    # Mix the inputs using lambda
    mixed_x = lam * x + (1 - lam) * x[index, :]

    #original and shuffled
    y_a, y_b = y, y[index]

    return mixed_x, y_a, y_b, lam

# Loss function using mixup
def mixup_criterion(criterion, pred, y_a, y_b, lam):

    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# mixup fit
def fit_with_mixup(model, train_loader, val_loader, optimizer, num_epochs=10, alpha=1.0):

    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        # Training loop with mixup
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            # Apply mixup to the batch
            inputs, targets_a, targets_b, lam = mixup_data(inputs, labels, alpha)

            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)

            # Compute mixup loss
            loss = mixup_criterion(loss_fn, outputs, targets_a, targets_b, lam)
            loss.backward()
            optimizer.step()

            # Calculate running loss and accuracy
            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            correct += (lam * predicted.eq(targets_a).sum().item() + (1 - lam) * predicted.eq(targets_b).sum().item())
            total += labels.size(0)

        # Calculate training loss and accuracy
        train_loss = running_loss / len(train_loader)
        train_acc = correct / total

        # Compute validation loss and accuracy (using compute_loss and compute_acc)
        val_loss = compute_loss(model, loss_fn, val_loader)
        val_acc = compute_acc(model, val_loader)

        # Store results in history
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_loss:.4f}, Train Acc: {train_acc*100:.2f}%")
        print(f"Validation Loss: {val_loss:.4f}, Validation Acc: {val_acc*100:.2f}%")

    return history


model = YourCNN(list_feature_maps=[32, 64], use_skip=True)
model.to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()


history = fit_with_mixup(model, train_loader, val_loader, optimizer, num_epochs=10, alpha=1.0)

# Evaluate on test data
test_acc = compute_acc(model, test_loader)
print(f"Test Accuracy with Mixup: {test_acc*100:.2f}%")
```

```python
# CutMix function
def cutmix_data(x, y, alpha=1.0):

    if alpha > 0:
        lam = np.random.beta(alpha, alpha)  # Sample lambda from beta distribution
    else:
        lam = 1

    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)  # Shuffle the batch

    # Get random box size for CutMix
    W, H = x.size(2), x.size(3)
    cut_rat = np.sqrt(1. - lam)
    cut_w = int(W * cut_rat)
    cut_h = int(H * cut_rat)

    # Randomly select box center
    cx = np.random.randint(W)
    cy = np.random.randint(H)

    # Ensure box boundaries don't exceed image size
    bbx1 = np.clip(cx - cut_w // 2, 0, W)
    bby1 = np.clip(cy - cut_h // 2, 0, H)
    bbx2 = np.clip(cx + cut_w // 2, 0, W)
    bby2 = np.clip(cy + cut_h // 2, 0, H)

    # Apply CutMix by replacing the region in x with another image
    x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]

    # Get two sets of labels
    y_a, y_b = y, y[index]
    # Adjust lambda based on the box size
    lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (W * H))
    return x, y_a, y_b, lam

# CutMix loss function
def cutmix_criterion(criterion, pred, y_a, y_b, lam):
    '''Calculate CutMix loss based on lambda'''
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

# Training function with CutMix
def fit_cutmix(model=None, train_loader=None, valid_loader=None, optimizer=None, num_epochs=50, alpha=1.0, verbose=True, seed=1234):
    torch.manual_seed(seed)
    model.to(device)

    if optimizer is None:
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    history = {'val_loss': [], 'val_acc': [], 'train_loss': [], 'train_acc': []}

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for X, y in train_loader:
            X, y = X.to(device), y.to(device)

            # Apply CutMix
            inputs, targets_a, targets_b, lam = cutmix_data(X, y, alpha=alpha)

            # Forward pass
            outputs = model(inputs.type(torch.float32))
            loss = cutmix_criterion(loss_fn, outputs, targets_a, targets_b, lam)

            # Backward and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        # Validation loss and accuracy
        val_loss = compute_loss(model, loss_fn, valid_loader)
        val_acc = compute_acc(model, valid_loader)
        train_loss = compute_loss(model, loss_fn, train_loader)
        train_acc = compute_acc(model, train_loader)

        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)

        if verbose:
            print(f"Epoch {epoch+1}/{num_epochs}")
            print(f"Train loss= {train_loss:.4f} - Train acc= {train_acc*100:.2f}% - Val loss= {val_loss:.4f} - Val acc= {val_acc*100:.2f}%")

    return history

# Initialize and train model with CutMix
model = YourCNN(list_feature_maps=[32, 64], use_skip=True)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Train model with CutMix
history_cutmix = fit_cutmix(model=model, train_loader=train_loader, valid_loader=val_loader,
                            optimizer=optimizer, num_epochs=10, alpha=1.0, verbose=True)

# Evaluate the model on the test set
test_accuracy = compute_acc(model, test_loader)

print(f"Test Accuracy of the CutMix Model: {test_accuracy*100:.2f}%")
```

<span style="color:red">**Question 3.5:**</span> Implement the **one-versus-all (OVA)** loss. The details are as follows:
- You need to apply `the sigmoid activation function` to logits $h = [h_1, h_2,...,h_M]$ instead of `the softmax activation` function as usual to obtain $p = [p_1, p_2,...,p_M]$, meaning that $p_i = sigmoid(h_i), i=1,...,M$. Note that $M$ is the number of classes.
- Given a data example $x$ with the ground-truth label $y$, the idea is to maximize the likelihood $p_y$ and to minimize the likelihoods $p_i, i \neq y$. Therefore, the objective function is to find the model parameters to
  - $\max\left\{ \log p_{y}+\sum_{i\neq y}\log(1-p_{i})\right\}$ or equivalently $\min\left\{ -\log p_{y}-\sum_{i\neq y}\log(1-p_{i})\right\}$.
  - For example, if $M=3$ and $y=2$, you need to minimize $\min\left\{ -\log(1-p_{1})-\log p_{2}-\log(1-p_{3})\right\}$.

Compare the model trained with the OVA loss and the same model trained with the standard cross-entropy loss.

```python
import numpy as np
import os
import torch
import matplotlib.pyplot as plt

# PGD attack function
def pgd_attack(model, input_image, input_label=None,
               epsilon=0.0313,  # Epsilon set to match previous example
               num_steps=20,    # Number of iterations for the attack
               step_size=0.002,  # Step size for gradient ascent (matching previous alpha)
               clip_value_min=0.0,  # Minimum value for clamping
               clip_value_max=1.0):  # Maximum value for clamping


    if isinstance(input_image, np.ndarray):
        input_image = torch.tensor(input_image, requires_grad=True)

    if isinstance(input_label, np.ndarray):
        input_label = torch.tensor(input_label)

    # evaluation mode
    model.eval()

    # Create a copy of the input image and set it to require gradients
    adv_image = input_image.clone().detach().requires_grad_(True)

    # Random initialization around input_image within epsilon bounds
    random_noise = torch.FloatTensor(input_image.shape).uniform_(-epsilon, epsilon).to(device)
    adv_image = adv_image + random_noise
    adv_image = torch.clamp(adv_image, clip_value_min, clip_value_max).detach().requires_grad_(True)

    # If no input label is provided, use the model's prediction
    if input_label is None:
        output = model(input_image)
        input_label = torch.argmax(output, dim=1)

    # Perform PGD attack
    for _ in range(num_steps):
        adv_image.requires_grad_(True)  # Ensure requires_grad is True in each iteration
        output = model(adv_image)
        loss = nn.CrossEntropyLoss()(output, input_label)
        model.zero_grad()
        loss.backward()

        # Check if gradient is available before accessing 'data'
        if adv_image.grad is not None:
            gradient = adv_image.grad.data
            adv_image = adv_image + step_size * gradient.sign()
            adv_image = torch.clamp(adv_image, input_image - epsilon, input_image + epsilon)  # Clip to a valid boundary
            adv_image = torch.clamp(adv_image, clip_value_min, clip_value_max)  # Clip to a valid range
            adv_image = adv_image.detach()  # Detach to prevent gradient accumulation
        else:
            print("Warning: Gradient is None. Check for detach operations.")

    return adv_image.detach()

# Evaluate robust accuracy (PGD or FGSM attack)
def evaluate_robust_accuracy(model, loader, eps=0.0313, alpha=0.002, num_iter=20, attack_fn=None):
    model.eval()
    correct = 0
    total = 0
    for images, labels in loader:
        images, labels = images.to(device), labels.to(device)

        # Apply the attack function: PGD or FGSM
        if attack_fn is not None:
            if attack_fn == fgsm_attack:  # FGSM doesn't need num_steps and step_size
                adv_images = attack_fn(model, images, labels, epsilon=eps)
            else:  # PGD uses num_steps and step_size
                adv_images = attack_fn(model, images, labels, epsilon=eps, num_steps=num_iter, step_size=alpha)
        else:
            adv_images = pgd_attack(model, images, labels, epsilon=eps, num_steps=num_iter, step_size=alpha)

        # Make predictions on adversarial images
        outputs = model(adv_images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    return correct / total

# Visualize original and adversarial images
def visualize_attack(model, loader, eps=0.0313, alpha=0.002, num_iter=20, num_images=20):
    model.eval()
    data_iter = iter(loader)


    images, labels = next(data_iter)
    images, labels = images.to(device), labels.to(device)


    adv_images = pgd_attack(model, images, labels, epsilon=eps, num_steps=num_iter, step_size=alpha)


    fig, axs = plt.subplots(num_images, 2, figsize=(15, 40))

    for i in range(num_images):
        # Clip images to [0, 1] range
        original_image = torch.clamp(images[i].detach().cpu(), 0, 1).permute(1, 2, 0).numpy()
        adversarial_image = torch.clamp(adv_images[i].detach().cpu(), 0, 1).permute(1, 2, 0).numpy()

        # Show original image
        axs[i, 0].imshow(original_image)
        axs[i, 0].set_title(f"Original: {labels[i].item()}", fontsize=10)  # Adjust fontsize
        axs[i, 0].axis('off')

        # Show adversarial image
        axs[i, 1].imshow(adversarial_image)
        axs[i, 1].set_title(f"Adversarial", fontsize=10)  # Adjust fontsize
        axs[i, 1].axis('off')

    # Automatically adjust subplot parameters to give some padding
    plt.tight_layout(pad=2.0)
    plt.show()

# Ensure best_model is loaded from file if available, or train it if necessary
def load_or_train_best_model(train_loader, val_loader, num_epochs=10, model_path='best_model.pth'):
    # Check if the best model file exists
    if os.path.exists(model_path):
        print(f"Loading the best model from '{model_path}'")
        best_model = YourCNN(list_feature_maps=[16, 32, 64], drop_rate=0.2, batch_norm=True, use_skip=True)
        best_model.load_state_dict(torch.load(model_path))  # Load saved model state
        best_model.to(device)
        best_model.eval()  # Set model to evaluation mode
    else:
        print(f"No model found at '{model_path}', training a new model...")
        best_model = YourCNN(list_feature_maps=[16, 32, 64], drop_rate=0.2, batch_norm=True, use_skip=True)
        best_model.to(device)

        # Train the model
        optimizer = torch.optim.Adam(best_model.parameters(), lr=0.001)
        fit(best_model, train_loader, val_loader, optimizer, num_epochs=num_epochs)

        # Save the trained model
        torch.save(best_model.state_dict(), model_path)
        print(f"Model trained and saved as '{model_path}'.")

    return best_model


best_model = load_or_train_best_model(train_loader, val_loader, num_epochs=10)

# Evaluate the model's robust accuracy under PGD attack
robust_accuracy = evaluate_robust_accuracy(best_model, test_loader, eps=0.0313, alpha=0.002, num_iter=20)
print(f"Robust Accuracy with PGD Attack: {robust_accuracy * 100:.2f}%")


visualize_attack(best_model, test_loader, eps=0.0313, alpha=0.002, num_iter=20, num_images=20)
```
